netter Chatbot 😇

```python
def netter_chatbot():

  # Frage nach dem Befinden
  fragenett = input("Hallo! Ich bin der nette Chatbot. Wie geht es dir?💖")
  if(fragenett.find("gut")!=-1):
      print("Das freut mich!! 💘")
  elif(fragenett.find("schlecht")!=-1):
   print("Das tut mir sehr leid, hoffentlich geht es dir bald besser! ❣")
  else:
    print("Ich freue mich sehr darauf, jetzt mit dir zu sprechen!💟")

  # Frage nach dem Namen
  name = input("Wie heißt du?")
  if(name.find("Karen")!=-1):
    print("Wir haben keinen Manager, also ist dieser Chatbot definitiv nichts für sie.")
  else:
    print("Das ist ein sehr schöner Name, " + name +"!" + " Freut mich, dich kennenzulernen!💙")

  # Frage nach Geburtstag
  frage=int(input("In welchem Monat (geschrieben als Zahl) hast du Geburtstag?"))
  if(frage==today.month):
    frage=int(input("An welchem Tag hast du Geburtstag?"))
    if(frage==today.day):
     x="Happy birthday to you"
     print(x+"!")
     print(x+"!")
     print("Happy birthday, dear "+ name+"!")
     print(x+"!")
     print("🥳 🎈 🎉 🎁")
    else:
      print("Oh, du hast ja diesen Monat Geburtstag!🥳")
  else:
    print()


  # Frage nach Nutella und Butter
  nutella = input("Nutella mit oder ohne Butter?🥯")
  if(nutella.find("ohne")!=-1):
    print("Nein, das sehe ich anders,  " + nutella + " schmeckt doch gar nicht!🤢")
  elif(nutella.find("mit")!=-1):
      print("Ja, mit " + nutella + " hast du vollkommen Recht!")
  elif(nutella.find("Ohne")!=-1):
    print("Nein, das sehe ich anders,  " + nutella + " schmeckt doch gar nicht!🤢")
  else:
      print("Cool!	🙃")


  # Frage nach Müsli und Milch
  müsli = input("Kommt zuerst die Milch oder das Müsli in die Schüssel?🥣 ")
  if(müsli.find("Joghurt")!=-1):
    print("Wenn du meinst" )
  elif(müsli.find("Milch")!=-1):
    print("Nein, wie willst du da die Mengenverhältnisse erkennen?🤮")
  elif(müsli.find("milch")!=-1):
    print("Nein, wie willst du da die Mengenverhältnisse erkennen?🤮")
  elif(müsli.find("Müsli")!=-1):
    print("Ja, sehe ich genau so! " + müsli + " ist die einzig richtige Antwort!")
  elif(müsli.find("müsli")!=-1):
    print("Ja, sehe ich genau so! " + müsli + " ist die einzig richtige Antwort!")
  else:
    print("Auf jeden Fall, gute Antwort!🤗")
```

gemeiner Chatbot 😠

```python
def gemeinerBot():

  #fragt wie es einem geht
  frage = input("Hallo! Ich bin Gudrun. Wie geht es dir?😠 ")

  if(frage.find("gut")!=-1): # in der Antwort steckt das Wort "gut"
      print("Das ändere ich jetzt 🖕")

  elif(frage.find("schlecht")!=-1): # in deiner Antwort steckt das Wort "schlecht"
      print("Ich hoffe das bleibt so, du blödes Stück Dreck 💩")

  else: # du hast nichts mit "gut" oder "schlecht" geantwortet
     print("Es interessiert mich eigentlich gar nicht...🙄")

  if(frage.find("dir?")!=-1): # in der Antwort steckt eine Frage zurück
   print("Versuchs gar nicht erst")

  #fragt nach dem Namen

  found=False
  namenNett=["Amalia", "Seba", "Feli", "Anna", "Rebecca", "Nina","Julia"]

  frage=input("wie heißt du? ")
  if(frage.find("Karen")!=-1): # falls der Name Karen ist
   print("Niemand mag dich 🤬")
  else:
   for i in namenNett: # hier gehen wir alle Namen der Liste mit "netten Namen" durch
      if(frage.find(i)!=-1): # der eingegebene Name ist in der Liste der netten Namen
       print("Ich mag den Namen!!😊")
       found=True
       break
   if(found==False): # falls nicht
       print("🤮")
```

# Tierquiz 🐈🐘🐧🐪

Dieser Bot findet heraus, welches Tier am besten zu dir passt! Er zeigt dir sogar ein Bild des Tieres an.

```python
# hier laden wir die Tierbilder hoch
# dafür müssen wir die Bilder der Tiere mit den entsprechenden Dateinamen in unserem Drive haben
# funktioniert daher nur in Google Colab

from google.colab import files
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import drive


drive.mount('/content/drive')

tier_list = ["Loewe","Kamel" , "Elefant", "Erdmaennchen", "Hyaene", "Pinguin", "Eisbaer",
             "Robbe", "Schneeleopard", "Polarfuchs", "Affe", "Faultier", "Panda", "Krokodil",
             "Tukan", "Katze", "Meerschweinchen", "Hund", "Fisch", "Kaninchen", "Patrick"]
tier_dict={}

def upload_files():
  for tier in tier_list:
    with open(f"drive/MyDrive/{tier}.jpg", "rb") as f:
      tier_dict[tier] = f.read()

def show_image(tier:str):
  im = Image.open(BytesIO(tier_dict[tier]))
  plt.imshow(im)
  plt.axis(False)
  plt.show()

upload_files()
```
